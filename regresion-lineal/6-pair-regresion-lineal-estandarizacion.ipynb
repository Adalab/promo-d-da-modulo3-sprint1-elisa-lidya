
Los objetivos del pair programming de hoy son:
- Sobre las variables predictoras, en este caso deberéis estandarizar estas variables. De nuevo, usad el método que prefiráis.


```python
numericas = df_socio.select_dtypes(include = np.number)
numericas.drop('life_expectancy_at_birth_total_bc', axis = 1, inplace = True)
```

```python
scaler = StandardScaler()
scaler.fit(numericas)
vp_esc = scaler.transform(numericas)

numericas_estandarizadas = pd.DataFrame(vp_esc, columns = numericas.columns)
numericas_estandarizadas.head(3)
```

```python
df_socio[numericas.columns] = numericas_estandarizadas
```
