```python
# Establecemos la columna 'country' como nuestro índice, ya que no queremos perder esta información pero no se trata de una variable predictora realmente.
# ------------------------------

df_socio = df_socio.set_index("country")
```

```python
df_socio.dtypes
```

> Nuestras dos variables categóricas ('region' y 'quantity_of_physicians') tienen orden, de modo que realizamos el encoding con 

```python
# Hacemos primero el encoding de la variable referente a la cantidad de médicos.
# ------------------------------

diccionario_medicos = {'Muchos': 3, 'Medios': 2.5, 'Muy pocos': 0, 'Pocos': 1}
```

```python
df_socio["quantity_of_physicians"] = df_socio["quantity_of_physicians"].map(diccionario_medicos)
```

```python
# Por último hacemos el encoding de la variable 'region'.
# Decidimos realizar dos encodings diferentes y por tanto guardamos dos dataframes distintos.
# ------------------------------

diccionario_eli = {'SouthAmerica': 2, 'Oceania': 3, 'WesternEurope': 3, 'EasternEurope': 2,
       'SouthernEurope': 3, 'NorthernAmerica': 3, 'EasternAsia': 3, 'WesternAsia': 2,
       'NorthernEurope': 3, 'NorthernAfrica': 1, 'SouthernAsia': 1,
       'South-easternAsia': 2, 'CentralAmerica': 2, 'SouthernAfrica':0}

diccionario_lidya = {'SouthAmerica': 4, 'Oceania': 5, 'WesternEurope': 5, 'EasternEurope': 3,
       'SouthernEurope': 5, 'NorthernAmerica': 5, 'EasternAsia': 5, 'WesternAsia': 4,
       'NorthernEurope': 5, 'NorthernAfrica': 2, 'SouthernAsia': 3,
       'South-easternAsia': 3, 'CentralAmerica': 3, 'SouthernAfrica': 0}
```

```python
df_socio_eli = df_socio.copy()
df_socio_lidya = df_socio.copy()

df_socio_eli["region"] = df_socio_eli["region"].map(diccionario_eli)
df_socio_lidya["region"] = df_socio_lidya["region"].map(diccionario_lidya)
```
